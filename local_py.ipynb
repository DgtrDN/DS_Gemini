### Проект 8: Симуляция банковской системы

````python
# Напиши класс Account (упрощённый, без ООП пока) - используй словари
accounts = {}

def create_account(account_id, initial_balance=0):
    accounts[account_id] = {"balance": initial_balance, "transactions": []}

def deposit(account_id, amount):
    if account_id in accounts:
        accounts[account_id]["balance"] += amount
        accounts[account_id]["transactions"].append({"type": "deposit", "amount": amount})

def withdraw(account_id, amount):
    if account_id in accounts and accounts[account_id]["balance"] >= amount:
        accounts[account_id]["balance"] -= amount
        accounts[account_id]["transactions"].append({"type": "withdraw", "amount": amount})

# Напиши функции:
# - get_balance(account_id)
# - transfer(from_account, to_account, amount)
# - get_transaction_history(account_id)
# - get_accounts_info() - вся информация о счетах
````
